```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import RandomOverSampler
import matplotlib.pyplot as plt
```

```python
# Upload files
train_val_df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/INFO 617 Group Project Train Val (1).xlsx')
test_df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/INFO 617_Group Project Test Set.xlsx')

# Quick peek at data
print("Train/Validation Data:")
print(train_val_df.head())

print("\nTest Data:")
print(test_df.head())
```

```python
print("Train/Val Shape:", train_val_df.shape)
print("Test Shape:", test_df.shape)
```

```python
# ğŸ“Œ 1. IMPORT LIBRARIES

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns

```

```python
#  2. LOAD TRAIN-VALIDATION DATA

# Check basic info
print(train_val_df.head())
print(train_val_df.columns)
```

```python
#  3. PREPARE TRAIN-VALIDATION INPUTS

# Assuming columns are 'Sentence' and 'Label'
texts = train_val_df['Sentence'].astype(str).tolist()
labels = train_val_df['Label'].tolist()

# Lowercase and strip spaces
texts = [text.lower().strip() for text in texts]

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=5000)

# Train-Test Split (inside train_val set)
X_train, X_val, y_train, y_val = train_test_split(
    texts, labels, test_size=0.1, random_state=42, stratify=labels
)

X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_val_tfidf = tfidf_vectorizer.transform(X_val)
```

```python
#  4. TRAIN RANDOM FOREST

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_tfidf, y_train)
```

```python
#  5. VALIDATION PERFORMANCE (Train-Validation Split)

# Predict
y_pred_val_rf = rf_model.predict(X_val_tfidf)

# Accuracy
rf_val_acc = accuracy_score(y_val, y_pred_val_rf)
print(f"Random Forest Validation Accuracy: {rf_val_acc*100:.2f}%")

# Confusion Matrix
cm_val_rf = confusion_matrix(y_val, y_pred_val_rf)

plt.figure(figsize=(10,8))
sns.heatmap(cm_val_rf, annot=True, fmt='d', cmap='Blues')
plt.title('Random Forest Confusion Matrix (Validation)')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Classification Report
print(classification_report(y_val, y_pred_val_rf))
```

```python
#  6. LOAD TEST DATA


# Prepare Test inputs
test_texts = test_df['Sentence'].astype(str).tolist()
test_labels = test_df['Label'].tolist()

test_texts = [text.lower().strip() for text in test_texts]

# TF-IDF transform using the SAME vectorizer
X_test_tfidf = tfidf_vectorizer.transform(test_texts)
```

```python
#  7. TEST PERFORMANCE

# Predict on Test Set
y_pred_test_rf = rf_model.predict(X_test_tfidf)

# Accuracy
rf_test_acc = accuracy_score(test_labels, y_pred_test_rf)
print(f"Random Forest Test Accuracy: {rf_test_acc*100:.2f}%")

# Classification Report
print(classification_report(test_labels, y_pred_test_rf))
```
